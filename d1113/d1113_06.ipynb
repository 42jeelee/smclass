```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
### 학교별 인원을 원 그래프로 출력하시오.
count_df = df.groupby('학교')['이름'].count()

# explode = [0.1, 0]
wedgeprops = {'width': 0.6, 'edgecolor': 'white', 'linewidth': 4}

plt.pie(count_df, labels=count_df.index, \
        autopct='%.1f%%', startangle=90, \
        wedgeprops=wedgeprops)
plt.show()
```
