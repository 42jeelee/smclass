```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
## 원그래프
values = [30, 25, 20, 13, 10, 2]
explode = [0.1, 0, 0, 0, 0, 0]
labels = ['python', 'java', 'javascript', 'oracle', 'html', 'etc']
colors = ['plum', 'beige', 'skyblue', 'bisque', 'khaki', 'peachpuff']

plt.title('프로그램 사용률')

## pie함수 (데이터값), labels: 글자 출력
## explode: 원 그래프 분리, countclock: 시계 반대 방향, False: 시계 방향
## sstartangle = 90: 12시부터 시작됨.
## autopct: 데이터 값 출력
plt.pie(values, labels=labels, explode=explode, autopct='%d%%', counterclock=False, startangle=90, colors=colors)

plt.legend(loc=(1.2, 0.3), title='언어별 선호도')

plt.show()
```
