```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
### 남자 인구 현황 그래프 출력하시오.
df = pd.read_excel('연령별인구현황_202410.xlsx', skiprows=3, usecols='B,E:Y', nrows=1, index_col='행정기관')
df

### 여자 인구 현황 그래프 출력하시오.
df2 = pd.read_excel('연령별인구현황_202410.xlsx', skiprows=3, usecols='b,AB:AV', nrows=1, index_col='행정기관')
df2
```

```python
# df.columns
# df.iloc[0].str.replace(',', '').astype(int)
# df2.columns
# df2.iloc[0].str.replace(',', '').astype(int)

df2.columns = df.columns
```

```python
### 막대 그래프로 출력하시오.
## 1. 데이터값, 천단위 표시 제거, int타입 변경
## 2. plt.bar(x, y) 출력
plt.figure(figsize=(15, 10))

plt.xticks(rotation=45)

plt.barh(df.columns, df.iloc[0].str.replace(',','').astype(int), label='남자')
plt.barh(df2.columns, -(df2.iloc[0].str.replace(',','').astype(int)) * 2, label='여자')
plt.ylabel('연령대')
plt.xlabel('인구수')

plt.legend()

plt.show()
```
