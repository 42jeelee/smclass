```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# x, y축의 데이터
x = [1, 2, 3]     # 리스트, 튜플, DataFrame 데이터 - 리스트 형태
y = [2, 4, 8]

# x = [1, 2, 3, 4, 5, 6, 7, 8]
# x = df.index
# y = df['수학']

# 선 그래프 생성
plt.plot(x, y)

## 상단 타이틀
plt.title("라인 그래프 Line", size=30)

# 그래프 축 설명
# 그래프 X축 설명, color: 글자색 변경 / 색상 값, 16코드
plt.xlabel('X축 날짜', color='#662E93', loc='right')  # loc: left, center, right
plt.ylabel('Y축 인원', color='#F834DD', loc='top')    # loc: top, center, bottom

# 그래프 눈금 단위
plt.xticks([1, 2, 3, 4, 5])
plt.yticks([2, 4, 8])

plt.show()
```
