```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx')
```

```python
len(df.index)
```

그래프 grid: 안내선 출력

```python
plt.plot(df['이름'], df['키'])
plt.xticks(rotation=45)

# 그래프 안내선, axis: x, y, 축으로 선이 그려짐, linestyle: 선모양
# linewidth: 선 두께, color: 선 색상
plt.grid(axis='y', linestyle='--', linewidth=2, color='purple', alpha=0.2)
plt.grid(axis='x', alpha=0.2)

# 그래프 값 추가
for i in range(len(df.index)):
  plt.text(df['이름'][i], df['키'][i]+0.5, df['키'][i], ha='center')

plt.show()
```
