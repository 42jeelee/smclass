```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
## x, y축의 데이터
x = [1, 2, 3]
y = [2, 4, 8]

plt.figure(figsize=(7, 5))
plt.plot(x, y, label='범례데이터', marker='o')
## text(x좌표위치, y좌표위치, 데이터값), ha=글자를 가운데 정렬
# plt.text(x[0], y[0] + 0.2, y[0], ha='center')

for i in range(len(y)):
  plt.text(x[i] + 0.05, y[i] - 0.25, y[i], ha='center')

plt.title('라인그래프')
plt.legend()
plt.show()
```
