```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
import random

df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
# x축: 지원번호, Y축 국어
# 선 색상: red, marker 색상: yellow, 테두리: blue, 선타입 --

x = df.index
y = df['국어']
e = df['영어']
m = df['수학']

# 그래프 크기 조정, dpi : 그래프 해상도 설정: 해상도가 높을수록 이미지 크기 커짐.
plt.figure(figsize=(20, 5), facecolor='#1283AD') # x축으로 5, y축으로 20 크기조절

plt.plot(x, y, 'ro--', label='국어', mfc='yellow', mec='blue')
plt.plot(x, e, 'v-', label='영어', mfc='yellow', mec='blue')
plt.plot(x, m, 'x-.', label='수학', mfc='yellow', mec='blue')

plt.title('국어 성적', size=30)

plt.xlabel('번호', loc='right')
plt.ylabel('점수', loc='top')

plt.legend()
# plt.savefig('graph2.png', dpi=300)  # 이미지 저장

plt.show()
```
