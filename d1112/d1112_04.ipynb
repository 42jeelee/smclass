```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# x, y축의 데이터
x = [1, 2, 3]     # 리스트, 튜플, DataFrame 데이터 - 리스트 형태
y = [2, 4, 8]
yy = [1, 8, 3]

# 선 그래프 생성
# linewidth : 선 두께, color : 선 색상
# marker : 지점 표시 - o, x, v, markersize(ms) : 마커 크기, markeredgecolor(mec) : 마커 테두리선, markerfacecolor(mfc) : 마커 색상
# alpha : 투명도 설정, linestyle(ls) : 라인 스타일 : 점선, -- dash선, -. 선
# plt.plot(x, y, label='인원증가', linewidth=5, marker='o', markersize=10, markeredgecolor='red', markerfacecolor='yellow')
# plt.plot(x, yy, label='인원변동', marker='v', ms=10, mec='blue', mfc='yellow', alpha=0.4, ls='-.')

# b: 선 색상, v : 마커 표시, : 선 모양
plt.plot(x, yy, 'bv:', label='비용변동')

## 상단 타이틀
plt.title("라인 그래프 Line", size=30)

# 범례 : 그래프 선 설명문구, loc: 범례 위치 upper, center, lower, left, center, right
# plt.legend(loc="upper center")
plt.legend(loc=(0.3, 0.1))  # 좌표값으로 위치 설정 가능
```
