```python
import pandas as pd

data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2015, 2017, 2014, 2015, 2017, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
### 관객 수 * 10000 수정하고, 관객 수로 역순 정렬하시오.
df.columns
```

```python
df['관객 수'] * 10000
```

```python
# 개봉 연도에서 '년도'를 추가해서 출력하시오.
df['개봉 연도'].astype(str) + '년'
```

```python
### 관객 수 천 단위 표시해서 다시 출력하시오.
'$' + df['관객 수'].map('{:,}'.format)
df
```

```python
### $ 없애고, 천단위 표시 없애고, int타입으로 변경하시오.
# df['관객 수'] = df['관객 수'].str[1:].str.replace(',', '').astype(int)
# df
```

```python
## 추천 점수 컬럼을 추가 : 추천 점수 = (관객 수 * 평점) // 100
# 해서 추가하시오.
df['추천 점수'] = (df['관객 수'] * df['평점']) // 100
df
```

```python
# 연도 그룹핑을 해서 관객 수 출력하시오.
df.groupby('개봉 연도')[['개봉 연도', '관객 수']].value_counts()
```
