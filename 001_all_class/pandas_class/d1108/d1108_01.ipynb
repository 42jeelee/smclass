데이터 조건 검색

```python
# 조건식 생성
filt = df['키'] >= 185
filt
# 조건식 대입
df[filt]
df[~filt]     # not에 대한 결과값 출력
df.loc[filt]  # row
df.loc[filt, ['국어', '수학']]  # row
df.loc[filt, '키']
```

문자열 함수 : startwith, contains, replace, lower, upper

```python
# startwith : 해당 값으로 시작되는 단어 검색
filt = df['이름'].str.startswith('김')
df[filt]

# contains : 해당 값이 포함되어 있는 단어 검색
filt = df['이름'].str.contains('수')
df.loc[filt, ['이름', '학교', '키']]

# isin() : 포함이 되어 있는 지 확인
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs)
df[filt]

# lower() : 해당 값을 모두 소문자로 변환
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
# 포함이 되어 있으면 모두 검색
# NaN 데이터 처리를 하지 않으면 에러
langs = ['python', 'java']
filter = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
### j 가 포함되어 있는 정보 검색
filt = df['SW특기'].str.lower().str.contains('j', na=False)
df[filt]
```

```python
filt = df['SW특기'].str.lower().isin(['python', 'java'])
df[filt]
```
