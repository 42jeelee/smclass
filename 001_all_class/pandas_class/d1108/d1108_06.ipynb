```python
import pandas as pd
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df

# 학교 구로고 -> 가산고, 디지털고 -> 단지고 로 변경하시오.
# SW특기 소문자로 변경하시오.
# 4번 SW특기 -> Django
# 5번 SW특기 -> python
# 합계컬럼을 추가하시오.
# 결과컬럼을 추가해서 400점 이상은 pass, 아니면 fail 출력하시오.

filt = df['학교'] == '구로고'
df.loc[filt, '학교'] = '가산고'
filt = df['학교'] == '디지털고'
df.loc[filt, '학교'] = '단지고'

df.rename(columns={'SW특기': 'sw특기'}, inplace=True)

df.loc['4번', 'sw특기'] = 'Django'
df.loc['5번', 'sw특기'] = 'python'
df
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
filt = df['합계'] >= 400

df.loc[filt, '결과'] = 'pass'
df.loc[~filt, '결과'] = 'fail'
df
```
