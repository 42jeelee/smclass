```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

* drink.csv 파일 불러와서
* total_servings 컬럼을 추가, beer_servings, spirit_servings, wine_servings 합계
* country별 막대 그래프를 생성하시오.

```python
df = pd.read_csv('drinks.csv')

df['total_servings'] = df['beer_servings'] + df['spirit_servings'] + df['wine_servings']
df
```

```python
plt.figure(figsize=(100, 25))

plt.bar(df['country'], df['total_servings'])

plt.xticks(rotation=90)
plt.show()
```
