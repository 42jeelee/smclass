```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
```

```python
# 1명 - row 추가
df.loc[8] = ['평균', 2024, df['관객 수'].mean(), df['평점'].mean()]
df
```

```python
### 영화 관객 수를 막대 그래프로 출력하시오.

### 평균 막대 그래프를 추가하고, 평균 선을 그어서 .. 표시하시오.
## 이름: 평균, 년도 2024 관객 수: 평균관객 수, 평점: 평균 평점
## row 1개를 추가하시오.

plt.figure(figsize=(15, 10))

s_bar = plt.bar(df['영화'], df['관객 수'])
plt.plot(df['영화'], [df.loc[8, '관객 수']] * len(df['영화']), color='black', ls='-.')

plt.xticks(rotation=45)
plt.ylim(1000, 1800)

s_bar[len(df.index) - 1].set_color('red')

plt.show()
```
