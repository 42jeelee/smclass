```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
## 원그래프
values = [30, 25, 20, 13, 10, 2]
# values = [1, 1, 1, 1, 2, 1]
explode = [0.2, 0.1, 0, 0, 0, 0]
labels = ['python', 'java', 'javascript', 'oracle', 'html', 'etc']
colors = ['plum', 'beige', 'skyblue', 'bisque', 'khaki', 'peachpuff']

## edgecolor: 테두리, linewidth: 테두리 두께
wedgeprops = {'width': 0.6, 'edgecolor': 'black', 'linewidth': 2}

### 데이터 값 : 2% 이하의 값은 출력되지 않도록 함수 적용
### pct : values의 퍼센트 값 생성
# def custom_autopct(pct):
#   if pct >= 10:
#     return '%.1f%%' % pct
#   return ''

def custom_autopct(pct):
  return f'{pct:.1f}%' if pct >= 10 else ''

## wedgeprops : 중간 공백, pctdistance : 글자 위치
plt.pie(values, labels=labels, autopct=custom_autopct, colors=colors, \
        wedgeprops=wedgeprops, startangle=90, counterclock=False, pctdistance=0.7)

plt.title('프로그램 사용률')
plt.legend(loc=(1.2, 0.3), title='언어별 선호도')

plt.show()
```
