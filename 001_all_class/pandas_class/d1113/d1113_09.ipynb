```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
## 복합 그래프 - 여러개 그래프
## 2 X 2 그래프를 조합해서 사용
#
fig, axs = plt.subplots(2, 2, figsize=(15, 13))
# 메인 타이틀 생성
fig.suptitle('4개 복합 그래프')

# 4개 그래프: 1번째 axs[0, 0], [0, 1], [1, 0], [1, 1]
axs[0, 0].bar(df['이름'], df['국어'], label='국어')
axs[0, 0].set_title('1. 국어점수 그래프')
axs[0, 0].set(xlabel='이름', ylabel='점수')
# 배경색
axs[0, 0].set_facecolor('lightyellow')
# 범례
axs[0, 0].legend(loc='upper center')

for i in range(len(df.index)):
  axs[0, 0].text(df['이름'].iloc[i], df['국어'].iloc[i] + 1, df['국어'].iloc[i], ha='center')

## 2번째: 선 그래프
axs[0, 1].set_title('2. 영어, 수학 점수 그래프')

axs[0, 1].plot(df['이름'], df['영어'], label='영어')
axs[0, 1].plot(df['이름'], df['수학'], label='수학')

axs[0, 1].set(xlabel='이름', ylabel='점수')
axs[0, 1].legend()

for i in range(len(df.index)):
  axs[0, 1].text(df['이름'].iloc[i], df['영어'].iloc[i] + 1, df['영어'].iloc[i], ha='center')
  axs[0, 1].text(df['이름'].iloc[i], df['수학'].iloc[i] + 1, df['수학'].iloc[i], ha='center')

## 3번째: 가로바 그래프
axs[1, 0].set_title('3. 가로바 키 그래프')
axs[1, 0].barh(df['이름'], df['키'])

axs[1, 0].set_xlim(150, 210)

for i in range(len(df.index)):
  axs[1, 0].text(df['키'].iloc[i] + 3, df['이름'].iloc[i], df['키'].iloc[i])

## 4번째: 선 그래프
axs[1, 1].set_title('4. 사회점수 선 그래프')
axs[1, 1].plot(df['이름'], df['사회'], color='g', alpha=0.5)

for i in range(len(df.index)):
  axs[1, 1].text(df['이름'].iloc[i], df['사회'].iloc[i] + 1, df['사회'].iloc[i], ha='center')

plt.show()
```
