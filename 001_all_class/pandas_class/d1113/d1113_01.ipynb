```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win용
matplotlib.rcParams['font.size'] = 15                 # 글자 크기
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac용
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
plt.title('학생 성적 그래프')
plt.plot(df.index, df['국어'], label='국어')
plt.plot(df.index, df['영어'], label='영어')
plt.plot(df.index, df['수학'], label='수학')

for i in range(len(df.index)):
  plt.text(df.index[i], df['국어'].iloc[i], df['국어'].iloc[i], ha='center')
  plt.text(df.index[i], df['영어'].iloc[i], df['영어'].iloc[i], ha='center')
  plt.text(df.index[i], df['수학'].iloc[i], df['수학'].iloc[i], ha='center')

plt.legend()

plt.show()
```

```python
def n_func(x, st=0):
  if (x + st) - 30 < st:
    return st
  return (x + st) - 30

n_list = [30, 35, -31, 36, 37]
for i in n_list:
  print("데이터 값 :", n_func(i))
```

```python
# 누적 막대 그래프
plt.figure(figsize=(8, 4))
plt.title('학생 성적 그래프')

plt.bar(df['이름'], df['국어'], width=0.6, label='국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], width=0.6, label='영어')
plt.bar(df['이름'], df['수학'], bottom=df['국어'] + df['영어'], width=0.6, label='수학')

for i in range(len(df['이름'])):
  plt.text(df['이름'].iloc[i], n_func(df['국어'].iloc[i]), df['국어'].iloc[i], ha='center')
  plt.text(df['이름'].iloc[i], n_func(df['영어'].iloc[i], df['국어'].iloc[i]), df['영어'].iloc[i], ha='center')
  plt.text(df['이름'].iloc[i], n_func(df['수학'].iloc[i], df['국어'].iloc[i] + df['영어'].iloc[i]), df['수학'].iloc[i], ha='center')

plt.legend()

plt.show()
```

```python
idx = np.arange(len(df.index))
w = 0.25

plt.bar(idx - w, df['국어'], width=w)
plt.bar(idx, df['영어'], width=w)
plt.bar(idx + w, df['수학'], width=w)

plt.show()
```
