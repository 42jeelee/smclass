```python
c_list = list(df.columns)
c_list
```

```python
type(c_list[0:8])   # list
type(c_list[-1])    # str
[1, 2] + [3, 4]     # list 타입을 맞춰서 더하기를 해야 함
```

```python
# 컬럼순서 변경 : 컬럼의 타입을 일치한 후 더하기를 해주면 됨.
# 리스트 - 리스트(문자열) + 리스트(문자열)
df = df[c_list[0:8] + [c_list[-1], c_list[8]]]
df
```

```python
# 평균 컬럼을 넣어서, 이름 바로 뒤로 위치를 변경하시오.
# 없는 컬럼을 입력하면 추가, 있는 컬럼에 값을 입력하면 변경
df['평균'] = df['총합'] / 5

c_list = list(df.columns)
c_list

df = df[[c_list[0], c_list[-1]] + c_list[1:-1]]
df
```

```python
# 평균 80점 이상, 결과 컬럼을 추가해서 평균 80점 이상이면 합격, 불합격 출력하시오.
# 결과 컬럼 추가
# 필터 적용
# 합격, 불합격 입력 조건에 맞는 값을 결과 컬럼에 입력을 하려면, df.loc
filt = df['평균'] >= 80
df.loc[filt, '결과'] = '합격'
df.loc[~filt, '결과'] = '불합격'
df
```

```python
## 컬럼 삭제 : drop ( columns ), 적용 inplace = True
# df.drop(columns=['평균'], inplace=True)       # 2개 컬럼 삭제
df.drop(columns=['결과', '총합'], inplace=True) # 2개 컬럼 삭제
df
```

```python
## index 삭제 : drop ( index ), 적용 inplace = True
# df.drop(index=['1번'])  # index를 가지고 삭제
df.drop(index=['1번', '3번'])  # index를 가지고 삭제
# 수학 점수가 80점 이하 삭제

filt = df['수학'] < 80
df.drop(index=df[filt].index)
df
```

```python
###
# 1. 요일 컬럼 숫자 0 추가
df['요일'] = 0
# 2. 0-6 까지의 숫자를 랜덤으로 수정 df.loc
def func(value):
  return random.randint(1, 6)

df['요일'] = df['요일'].apply(func)
df
# 3. 함수를 사용해서 0-일요일, 1-월요일, 2-화요일 ... 6-토요일로 변경하시오.
def func2(value):
  DAYS = ['일요일', '월요일', '화요일', '수요일', '목요일', '금요일', '토요일']

  return DAYS[value]

df['요일'] = df['요일'].apply(func2)
df
```
